## Warm Up Task A

```python
def warmUp_task_A(inputString):
    evenPlaces = ''
    oddPlaces = ''
    for i in range(0, len(inputString)):
        if ((i+1) % 2 == 0):
            evenPlaces = evenPlaces + inputString[i].upper()
            oddPlaces = oddPlaces + inputString[i]
        else:
            evenPlaces = evenPlaces + inputString[i]
            oddPlaces = oddPlaces + inputString[i].upper()
    return(evenPlaces , oddPlaces)
```

### Testing task A

```python
warmUp_task_A('abcdef')
```

## Warm Up Task B

```python
def warmUp_task_B(inputString):
    tempDictionary = {}
    for tempCharacter in inputString.lower():
        try:
            tempDictionary[tempCharacter] = tempDictionary[tempCharacter] + 1
        except:
            tempDictionary[tempCharacter] = 0
    answer = 0
    for tempKey , tempValue in tempDictionary.items():
        if(tempValue > 0):
            answer = answer + 1
    return(answer)
```

### Testing task B

```python
warmUp_task_B('RhabarbArka')
```
